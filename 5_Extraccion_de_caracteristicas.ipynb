```python
Max=Micro_Retailer['_latitude'].max()
Min=Micro_Retailer['_latitude'].min()
Limites =[Min,Max]
Limites
```

```python
Rango=Max-Min
Rango
```

```python
#Regla de Sturges
ni= 1 + 3.32*np.log10(n)
ni
```

```python
#ancho del intervalo
i=Rango/ni
i
```

```python
intervalos=np.linspace(Min,Max,9)
intervalos
```
